```python
strings = ["apple", "banana", "cherry"]
upper_strings = [s.upper() for s in strings]
upper_strings
```

```python
# 4
nums = [-5, 3, -2, 8, 0, 7]
positives = [x for x in nums if x > 0]
positives
```

```python
# 5
tuples_list = [(x, x**2) for x in range(1, 6)]
tuples_list
```

```python
# 6
text = "hello world"
vowels = [ch for ch in text if ch.lower() in "aeiou"]
vowels
```

```python
# 7
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [item for row in matrix for item in row]
flat
```

```python
# 8
nums2 = [-3, 5, -1, 9]
replaced = [x if x >= 0 else 0 for x in nums2]
replaced
```

```python
# 9
words = ["apple", "dog", "banana"]
lengths = [len(word) for word in words]
lengths
```

```python
# 10
words2 = ["Apple", "banana", "Avocado", "grape"]
filtered = [word for word in words2 if not word.lower().startswith('a')]
filtered
```

```python
# 11
numbers = [1, 2, 3, 4, 5]
even_odd = ["even" if x % 2 == 0 else "odd" for x in numbers]
even_odd
```

```python
# 12
divisible = [x for x in range(1, 101) if x % 3 == 0 and x % 5 == 0]
divisible
```

```python
# 13
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
table
```

```python
# 14
d = {"a": 1, "b": 2, "c": 3}
keys_list = [key for key in d]
keys_list
```

```python
# 15
string = "a1b2c3d4"
digits = [ch for ch in string if ch.isdigit()]
digits
```

```python
# 16
string2 = "hello world python"
no_spaces = [ch for ch in string2 if ch != " "]
no_spaces
```

```python
# 17
string3 = "programming"
duplicates = [ch for ch in set(string3) if string3.count(ch) > 1]
```

```python
# 18
sentences = ["hello world", "python is fun"]
all_words = [word for sentence in sentences for word in sentence.split()]
all_words
```

```python
# 19
lst = [1, 2, 2, 3, 4, 4, 5]
unique = [x for i, x in enumerate(lst) if x not in lst[:i]]
```
