```python
# 2
is_even = lambda x: x % 2 == 0
is_even(64)
```

```python
# 3
last_char = lambda s: s[-1]
last_char("anshika")
```

```python
# 4
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, nums))
squared
```

```python
# 5
nums2 = [1, 2, 3, 4, 5]
odds = list(filter(lambda x: x % 2 != 0, nums2))
odds
```

```python
# 6
tuples_list = [(1, 3), (2, 1), (4, 2)]
sorted_tuples = sorted(tuples_list, key=lambda x: x[1])
sorted_tuples
```

```python
# 7
is_palindrome = lambda s: s == s[::-1]
is_palindrome('madam')
```

```python
# 8
max_three = lambda a, b, c: max(a, b, c)
max_three(3,6,8)
```

```python

# 9
reverse = lambda s: s[::-1]
reverse("anshika")
```

```python
# 10
str_nums = ["1", "2", "3"]
int_nums = list(map(lambda x: int(x), str_nums))
int_nums
```

```python
# 11
strings = ["hello", "", "world", ""]
non_empty = list(filter(lambda x: x != "", strings))
non_empty
```

```python
# 12
from functools import reduce
factorial = lambda n: reduce(lambda x, y: x * y, range(1, n + 1))
factorial(4)
```

```python
# 13
larger = lambda a, b: a if a > b else b
larger(23,67)
```

```python
# 14
div_by_5 = lambda x: x % 5 == 0
div_by_5(50)
```

```python
# 15
nums3 = [5, 10, 15]
added = list(map(lambda x: x + 10, nums3))
added
```

```python
# 16
people = [{"name": "A", "age": 25}, {"name": "B", "age": 20}]
sorted_people = sorted(people, key=lambda x: x["age"])
sorted_people
```

```python
# 17
is_vowel = lambda ch: ch.lower() in "aeiou"
is_vowel('f')
```

```python
# 18
words = ["python", "java", "developer", "code"]
long_words = list(filter(lambda x: len(x) > 5, words))
long_words
```
