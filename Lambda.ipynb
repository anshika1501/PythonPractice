```python
# Book price data by standard
books = {
    "Hindi":    {"1-4": 60,  "5-8": 100, "9-10": 150},
    "Marathi":  {"1-4": 60,  "5-8": 100, "9-10": 150},
    "English":  {"1-4": 80,  "5-8": 100, "9-10": 150},
    "Science":  {"1-4": 90,  "5-8": 120, "9-10": 200},
    "Maths":    {"1-4": 100, "5-8": 140, "9-10": 250}
}

# Notebook prices
notebooks = {
    "Square": {"100_pages": 40, "200_pages": 70},
    "4_lines": {"100_pages": 30, "200_pages": 50},
    "2_lines": {"100_pages": 30, "200_pages": 50},
    "Single_lines": {"100_pages": 60, "200_pages": 100},
    "A4": {"100_pages": 100, "200_pages": 180}
}

# Print books table
print("BOOK PRICES")
print("Book\t\t1-4\t5-8\t9-10")
for book, price in books.items():
    print(f"{book}\t\t{price['1-4']}\t{price['5-8']}\t{price['9-10']}")

print("\nNOTEBOOK PRICES")
print("Type\t\t100 pages\t200 pages")
for nb, price in notebooks.items():
    print(f"{nb}\t\t{price['100_pages']}\t\t{price['200_pages']}")
```

```python
n = int(input("Enter number of items: "))

bill = {}   # item : [quantity, price]

for i in range(n):
    item = input("Enter item name: ")
    qty = int(input("Enter quantity: "))
    price = float(input("Enter price per item: "))
    bill[item] = [qty, price]   # âœ… inside loop

def printbill():
    print("\n" + "-"*40)
    print("{:^40}".format("BILL"))
    print("-"*40)

    print("{:<15}{:<10}{:<10}{:<10}".format("Item", "Qty", "Price", "Total"))
    print("-"*40)

    grand_total = 0

    for item, values in bill.items():
        qty, price = values
        total = qty * price
        grand_total += total
        print("{:<15}{:<10}{:<10}{:<10}".format(item, qty, price, total))

    print("-"*40)
    print("{:<35}{:<10}".format("Grand Total", grand_total))
    print("-"*40)

decide = input("Do you want to print the bill? Y/N: ")

if decide.lower() == 'y':
    printbill()
else:
    print("Thank you for visiting")
```
